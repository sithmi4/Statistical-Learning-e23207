#Q1) Prediction of fraudulent transactions

A bank monitors credit-card transactions for possible fraud.

For each transaction, let the sample space be

$$
\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\},
$$

where:

* $\omega_1$: low-risk customer, legitimate transaction,
* $\omega_2$: low-risk customer, fraudulent transaction,
* $\omega_3$: high-risk customer, legitimate transaction,
* $\omega_4$: high-risk customer, fraudulent transaction.

Assume the probabilities are

$$
P(\{\omega_1\})=0.50,\qquad
P(\{\omega_2\})=0.10,\qquad
P(\{\omega_3\})=0.20,\qquad
P(\{\omega_4\})=0.20.
$$

Let $X:\Omega\to\mathbb R$ denote the fraud indicator:

$$
X(\omega_1)=0,\qquad
X(\omega_2)=1,\qquad
X(\omega_3)=0,\qquad
X(\omega_4)=1.
$$

Thus $X=1$ means the transaction is fraudulent, and $X=0$ means it is legitimate.

Now suppose the bank does not observe the full outcome $\omega$, but only the customer's risk class. Define $Y:\Omega\to\mathbb R$ by

$$
Y(\omega_1)=0,\qquad
Y(\omega_2)=0,\qquad
Y(\omega_3)=1,\qquad
Y(\omega_4)=1.
$$

Thus:

* $Y=0$ means “low-risk customer,”
* $Y=1$ means “high-risk customer.”

Answer the following.

1. Write down the probability space $(\Omega,\mathfrak F,P)$, where $\mathfrak F=\mathfrak P(\Omega)$.

2. Compute the $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$.

3. Explain why $E[X\mid Y]$ must be $\sigma(Y)$-measurable, and therefore must be constant on the sets $\{ \omega_1,\omega_2\}$ and $\{\omega_3,\omega_4\}$.

4. Compute $E[X\mid Y]$ explicitly and show that

$$
E[X\mid Y]= \frac{1}{6} \mathbb{1}_{\{Y=0\}} + \frac{1}{2} \mathbb{1}_{\{Y=1\}}
$$

5. Verify the defining property of conditional expectation by checking that for every set $A\in\sigma(Y)$,

$$
\int_A E[X\mid Y]\,dP=\int_A X\,dP.
$$

6. Find the Conditional PMF $p_{X|Y}(x \mid y)$.

7. Verify the identity:
$$\mathbb{E}[X\mid Y=y] = \sum_{x \in \{0,1\}} x \cdot p_{X|Y}(x \mid y)$$.

8. Calculate the **Mean Square Error (MSE)**

9. How would the **Mean Square Error (MSE)** change if the bank decided to ignore the customer risk class $Y$ and instead used the global average fraud rate ($E[X] = 0.30$) as a constant estimate for every transaction? Calculate this error and explain its relationship to the **Law of Total Variance**.

10. Using the joint distribution of $(X, Y)$ derived from the sample space:

    a) Compute the entropy $H(X)$, representing the bank's initial uncertainty regarding fraud.

    b) Compute the conditional entropy $H(X \mid Y)$, representing the uncertainty remaining after the customer's risk class is revealed.

    c) Calculate the Mutual Information $I(X; Y)$.

    d) Based on these values, what percentage of the total uncertainty about fraud is removed by knowing the customer's risk class?

```python


import numpy as np
import math

Omega = ['ω1', 'ω2', 'ω3', 'ω4']

P = {
    'ω1': 0.50,   # low-risk, legitimate
    'ω2': 0.10,   # low-risk, fraudulent
    'ω3': 0.20,   # high-risk, legitimate
    'ω4': 0.20    # high-risk, fraudulent
}


X = {
    'ω1': 0,
    'ω2': 1,
    'ω3': 0,
    'ω4': 1
}


Y = {
    'ω1': 0,
    'ω2': 0,
    'ω3': 1,
    'ω4': 1
}

print("="*60)
print("1. Probability Space (Ω, F, P)")
print("="*60)
print("Ω =", Omega)
print("F = Power set of Ω")
print("P =", P)


# 2. Sigma Algebra Generated by Y


sigma_Y = [
    set(),
    {'ω1', 'ω2'},        # Y = 0
    {'ω3', 'ω4'},        # Y = 1
    set(Omega)
]

print("\n" + "="*60)
print("2. σ(Y)")
print("="*60)
for s in sigma_Y:
    print(s)

# 3. Conditional Expectation E[X|Y]


# E[X|Y=0]
P_Y0 = P['ω1'] + P['ω2']
EX_given_Y0 = (X['ω1']*P['ω1'] + X['ω2']*P['ω2']) / P_Y0

# E[X|Y=1]
P_Y1 = P['ω3'] + P['ω4']
EX_given_Y1 = (X['ω3']*P['ω3'] + X['ω4']*P['ω4']) / P_Y1

# Define E[X|Y](ω)
EX_given_Y = {}
for w in Omega:
    if Y[w] == 0:
        EX_given_Y[w] = EX_given_Y0
    else:
        EX_given_Y[w] = EX_given_Y1

print("\n" + "="*60)
print("3. Conditional Expectation E[X|Y]")
print("="*60)
print(f"E[X|Y=0] = {EX_given_Y0:.6f} = 1/6")
print(f"E[X|Y=1] = {EX_given_Y1:.6f} = 1/2")
print("\nHence:")
print("E[X|Y] = (1/6)·1{Y=0} + (1/2)·1{Y=1}")


# 4. Verify Conditional Expectation Property


print("\n" + "="*60)
print("4. Verification of Conditional Expectation Property")
print("="*60)

for A in sigma_Y:
    lhs = sum(EX_given_Y[w] * P[w] for w in A)
    rhs = sum(X[w] * P[w] for w in A)
    print(f"A = {A}")
    print(f"∫A E[X|Y] dP = {lhs:.6f}")
    print(f"∫A X dP      = {rhs:.6f}")
    print()

# 5. Conditional PMF p(X|Y)


print("\n" + "="*60)
print("5. Conditional PMF p(X|Y)")
print("="*60)

# For Y = 0
p_X_given_Y = {
    0: {0: 5/6, 1: 1/6},   # Y=0: P(X=0|Y=0), P(X=1|Y=0)
    1: {0: 1/2, 1: 1/2}    # Y=1
}

for y in [0, 1]:
    print(f"Y = {y}")
    for x in [0, 1]:
        print(f"  P(X={x} | Y={y}) = {p_X_given_Y[y][x]:.6f}")


# 6. Verify E[X|Y=y] = Σ x p(X|Y)


print("\n" + "="*60)
print("6. Verification Using Conditional PMF")
print("="*60)

for y in [0, 1]:
    expected = sum(x * p_X_given_Y[y][x] for x in [0, 1])
    print(f"E[X|Y={y}] = {expected:.6f}")

# 7. Mean Square Error (MSE) Using E[X|Y]


# MSE = E[(X - E[X|Y])²]
MSE = sum((X[w] - EX_given_Y[w])**2 * P[w] for w in Omega)

print("\n" + "="*60)
print("7. Mean Square Error (Using E[X|Y])")
print("="*60)
print(f"MSE = E[(X - E[X|Y])²] = {MSE:.6f}")


# 8. Constant Estimator Using Global Mean E[X]


EX = sum(X[w] * P[w] for w in Omega)
MSE_constant = sum((X[w] - EX)**2 * P[w] for w in Omega)

print("\n" + "="*60)
print("8. Constant Estimator Using Global Mean")
print("="*60)
print(f"E[X] = {EX:.6f}")
print(f"MSE = E[(X - E[X])²] = {MSE_constant:.6f}")


Var_EX_given_Y = sum(
    (EX_given_Y0 - EX)**2 * P_Y0 +
    (EX_given_Y1 - EX)**2 * P_Y1
    for _ in [0]
)

print(f"Var(E[X|Y]) = {Var_EX_given_Y:.6f}")
print(f"Check: Var(X) = MSE + Var(E[X|Y])")
print(f"{MSE_constant:.6f} = {MSE:.6f} + {Var_EX_given_Y:.6f}")


# 9. Joint Distribution of (X, Y)

joint = {
    (0, 0): P['ω1'],
    (1, 0): P['ω2'],
    (0, 1): P['ω3'],
    (1, 1): P['ω4']
}

print("\n" + "="*60)
print("9. Joint Distribution p(X,Y)")
print("="*60)
for key, value in joint.items():
    print(f"P(X={key[0]}, Y={key[1]}) = {value:.6f}")


# 10. Entropy Function

def entropy(probabilities):
    h = 0
    for p in probabilities:
        if p > 0:
            h -= p * math.log2(p)
    return h


# 11. Entropy H(X)


P_X0 = P['ω1'] + P['ω3']  # 0.70
P_X1 = P['ω2'] + P['ω4']  # 0.30

H_X = entropy([P_X0, P_X1])

print("\n" + "="*60)
print("10(a). Entropy H(X)")
print("="*60)
print(f"P(X=0) = {P_X0:.6f}")
print(f"P(X=1) = {P_X1:.6f}")
print(f"H(X) = {H_X:.6f} bits")

# 12. Conditional Entropy H(X|Y)

H_X_given_Y0 = entropy([5/6, 1/6])
H_X_given_Y1 = entropy([1/2, 1/2])

H_X_given_Y = P_Y0 * H_X_given_Y0 + P_Y1 * H_X_given_Y1

print("\n" + "="*60)
print("10(b). Conditional Entropy H(X|Y)")
print("="*60)
print(f"H(X|Y=0) = {H_X_given_Y0:.6f} bits")
print(f"H(X|Y=1) = {H_X_given_Y1:.6f} bits")
print(f"H(X|Y)   = {H_X_given_Y:.6f} bits")


# 13. Mutual Information I(X;Y)


I_XY = H_X - H_X_given_Y

print("\n" + "="*60)
print("10(c). Mutual Information I(X;Y)")
print("="*60)
print(f"I(X;Y) = {I_XY:.6f} bits")

# 14. Percentage of Uncertainty Removed


percentage = (I_XY / H_X) * 100

print("\n" + "="*60)
print("10(d). Percentage of Uncertainty Removed")
print("="*60)
print(f"Percentage = {percentage:.2f}%")



print("\n" + "="*60)
print("FINAL ANSWERS SUMMARY")
print("="*60)
print("E[X|Y] = (1/6)·1{Y=0} + (1/2)·1{Y=1}")
print(f"MSE using E[X|Y]      = {MSE:.6f}")
print(f"MSE using E[X]=0.30   = {MSE_constant:.6f}")
print(f"H(X)                  = {H_X:.6f} bits")
print(f"H(X|Y)                = {H_X_given_Y:.6f} bits")
print(f"I(X;Y)                = {I_XY:.6f} bits")
print(f"Uncertainty Removed   = {percentage:.2f}%")
```

# Q2) Conditional Density of the Multivariate Gaussian

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.

1. Show that the conditional distribution is given by $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right),$$
where
* $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$

```python


import numpy as np
from scipy.stats import multivariate_normal
from numpy.linalg import inv, det

# 1. DEFINE DIMENSIONS


n = 2
m = 1


mu_X = np.array([1.0, 2.0])
mu_Y = np.array([3.0])

mu = np.concatenate([mu_X, mu_Y])

print("="*60)
print("MEAN VECTOR")
print("="*60)
print("μ_X =", mu_X)
print("μ_Y =", mu_Y)
print("μ   =", mu)

# 3. DEFINE BLOCK COVARIANCE MATRIX

Sigma_XX = np.array([
    [4.0, 1.0],
    [1.0, 3.0]
])

Sigma_XY = np.array([
    [1.0],
    [2.0]
])

Sigma_YX = Sigma_XY.T

Sigma_YY = np.array([
    [5.0]
])

Sigma = np.block([
    [Sigma_XX, Sigma_XY],
    [Sigma_YX, Sigma_YY]
])

print("\n" + "="*60)
print("COVARIANCE MATRIX")
print("="*60)
print("Σ =")
print(Sigma)

print("\nDeterminant |Σ| =", det(Sigma))

# 4. CONDITIONAL MEAN AND COVARIANCE


Sigma_YY_inv = inv(Sigma_YY)

Sigma_X_given_Y = (
    Sigma_XX
    - Sigma_XY @ Sigma_YY_inv @ Sigma_YX
)

print("\n" + "="*60)
print("CONDITIONAL COVARIANCE")
print("="*60)
print("Σ_X|Y =")
print(Sigma_X_given_Y)


y = np.array([4.0])

mu_star = (
    mu_X
    + Sigma_XY @ Sigma_YY_inv @ (y - mu_Y)
).flatten()

print("\n" + "="*60)
print("CONDITIONAL MEAN")
print("="*60)
print("Observed y =", y)
print("μ* =")
print(mu_star)

# 5. DEFINE CONDITIONAL DENSITY FUNCTION

def conditional_density(x):
    """
    Computes f_{X|Y}(x | y)
    where X|Y=y ~ N(mu_star, Sigma_X_given_Y)
    """
    return multivariate_normal.pdf(
        x,
        mean=mu_star,
        cov=Sigma_X_given_Y
    )

# 6. EVALUATE CONDITIONAL DENSITY AT SAMPLE POINTS

test_points = [
    np.array([1.0, 2.0]),
    np.array([1.2, 2.4]),
    np.array([0.0, 0.0]),
    mu_star
]

print("\n" + "="*60)
print("CONDITIONAL DENSITY VALUES")
print("="*60)

for x in test_points:
    fx = conditional_density(x)
    print(f"x = {x}")
    print(f"f_X|Y(x|y) = {fx:.8f}\n")

# 7. VERIFY AGAINST JOINT / MARGINAL RATIO

joint_dist = multivariate_normal(mean=mu, cov=Sigma)
marginal_Y = multivariate_normal(mean=mu_Y, cov=Sigma_YY)

print("\n" + "="*60)
print("VERIFY: f_X|Y = f_X,Y / f_Y")
print("="*60)

for x in test_points:
    z = np.concatenate([x, y])

    joint_pdf = joint_dist.pdf(z)
    marginal_pdf = marginal_Y.pdf(y)
    ratio = joint_pdf / marginal_pdf
    direct = conditional_density(x)

    print(f"x = {x}")
    print(f"Direct conditional pdf : {direct:.8f}")
    print(f"Joint/Marginal ratio   : {ratio:.8f}")
    print(f"Difference             : {abs(direct-ratio):.12e}\n")


# 8. DISPLAY THEORETICAL FORMULA

print("\n" + "="*60)
print("THEORETICAL RESULT")
print("="*60)
print("X | Y = y ~ N( μ*, Σ_X|Y )")
print()
print("μ* = μ_X + Σ_XY Σ_YY^{-1}(y - μ_Y)")
print("Σ_X|Y = Σ_XX - Σ_XY Σ_YY^{-1} Σ_YX")

# 9. OPTIONAL MONTE CARLO VERIFICATION

N = 100000
samples = np.random.multivariate_normal(mu, Sigma, size=N)

X_samples = samples[:, :n]
Y_samples = samples[:, n:]

tolerance = 0.05
mask = np.abs(Y_samples[:, 0] - y[0]) < tolerance

selected_X = X_samples[mask]

print("\n" + "="*60)
print("MONTE CARLO VERIFICATION")
print("="*60)
print(f"Observed y = {y[0]}")
print(f"Tolerance  = ±{tolerance}")
print(f"Selected samples = {len(selected_X)}")

if len(selected_X) > 10:
    empirical_mean = np.mean(selected_X, axis=0)
    empirical_cov = np.cov(selected_X.T)

    print("\nEmpirical conditional mean:")
    print(empirical_mean)

    print("\nTheoretical conditional mean:")
    print(mu_star)

    print("\nEmpirical conditional covariance:")
    print(empirical_cov)

    print("\nTheoretical conditional covariance:")
    print(Sigma_X_given_Y)
else:
    print("Not enough selected samples.")
```

# Q3) Continuous Fault Monitoring of a Spindle Motor

A factory monitors the health of a high-speed CNC spindle. The critical variable is the **Internal Temperature ($X$)**, but because the motor is sealed, they can only observe the **Total Spindle Power ($Y$)**.

**The Setup:**
1.  **Hidden Temperature ($X$):** The standardized temperature deviation follows a standard normal distribution:
    $$X \sim \mathscr{N}(0, 1)$$
2.  **Mechanical Load ($Q$):** The torque required to cut material creates "noise" in the power reading. It is independent of temperature and follows:
    $$Q \sim \mathscr{N}(0, 4)$$
3.  **The Measurement ($Y$):** The power sensor $Y$ measures the combined effect of thermal load and mechanical torque:
$$Y = X + Q$$

---

**Answer the followings:**

**1. Joint Distribution**
Identify the distribution of the random varaible $(X, Y)$.
* Show that $Y \sim \mathscr{N}(0, 5)$.
* Find the covariance $\text{Cov}(X, Y)$.

**2. Information Geometry**
Explain why $X$ is **not** $\sigma(Y)$-measurable. In practical terms, if the power meter shows $Y = 5$, why can't the engineer know for certain if the motor is overheating ($X$ is high) or if the material being cut is simply very hard ($Q$ is high)?

**3. Conditional Expectation (The Best Estimate)**
Compute the conditional expectation $E[X \mid Y]$.
* Using the property for joint normals: $E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$.
* Express this as a "signal-to-total-variance" ratio.

**4. Verification of the Residual**
Let $\widehat{X} = E[X \mid Y]$. Define the estimation error (residual) as $Z = X - \widehat{X}$.
* Prove that $E[Z] = 0$.
* Verify that the error $Z$ is uncorrelated with the observation $Y$ (the Orthogonality Principle).

**5. Information Theory: Differential Entropy**
* Compute the differential entropy $h(X)$.
* Compute the conditional differential entropy $h(X \mid Y)$.
* *Note: The variance of the conditional distribution $(X \mid Y=y)$ is given by $\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)}$.*

**6. Mutual Information**
Calculate $I(X; Y) = h(X) - h(X \mid Y)$.
* What percentage of the "uncertainty" about the temperature is removed by observing the power meter?
* How would this change if the mechanical noise $Q$ had a variance of $100$ instead of $4$?

**7. Diagnostic Interpretation**
The factory manager says: *"Since $Y = X + Q$, our best estimate of temperature should just be $Y$ itself."*
Using your result from Question 3, explain why $Y$ is a **biased** estimator of $X$ and why $E[X \mid Y]$ "shrinks" the measurement toward the mean (zero).



```python

# X ~ N(0,1)      : Internal Temperature (hidden)
# Q ~ N(0,4)      : Mechanical Load Noise (independent of X)
# Y = X + Q       : Observed Total Spindle Power


import numpy as np
import math
# 1. PARAMETERS

mu_X = 0
var_X = 1
sigma_X = np.sqrt(var_X)

mu_Q = 0
var_Q = 4
sigma_Q = np.sqrt(var_Q)

# Since Y = X + Q
mu_Y = mu_X + mu_Q
var_Y = var_X + var_Q
sigma_Y = np.sqrt(var_Y)

# Covariance
cov_XY = var_X       # Cov(X, X+Q) = Var(X) = 1

# Correlation
corr_XY = cov_XY / (sigma_X * sigma_Y)

# Joint mean vector and covariance matrix
mu = np.array([mu_X, mu_Y])

Sigma = np.array([
    [var_X, cov_XY],
    [cov_XY, var_Y]
])

print("="*70)
print("1. JOINT DISTRIBUTION OF (X, Y)")
print("="*70)
print("(X, Y) is jointly Gaussian:")
print("(X, Y) ~ N(mu, Sigma)")
print()
print("mu =")
print(mu)
print()
print("Sigma =")
print(Sigma)
print()
print("Therefore:")
print("Y ~ N(0, 5)")
print(f"Cov(X, Y) = {cov_XY}")
print(f"Corr(X, Y) = {corr_XY:.6f}")

# 2. INFORMATION GEOMETRY

print("\n" + "="*70)
print("2. INFORMATION GEOMETRY")
print("="*70)
print("X is NOT σ(Y)-measurable because Y = X + Q and Q is random.")
print("Knowing Y does not uniquely determine X.")
print()
print("Example: If Y = 5")
print("  - X = 5, Q = 0")
print("  - X = 3, Q = 2")
print("  - X = 1, Q = 4")
print("  - X = 7, Q = -2")
print("All are possible.")
print("Hence the engineer cannot know the exact temperature.")

# 3. CONDITIONAL EXPECTATION
# E[X|Y=y] = Cov(X,Y)/Var(Y) * y = (1/5)y

gain = cov_XY / var_Y     # 1/5

print("\n" + "="*70)
print("3. CONDITIONAL EXPECTATION (BEST ESTIMATE)")
print("="*70)
print("E[X | Y=y] = Cov(X,Y)/Var(Y) * y")
print(f"            = {cov_XY}/{var_Y} * y")
print(f"            = {gain:.6f} y")
print()
print("Therefore:")
print("E[X | Y] = (1/5) Y")
print()
print("Signal-to-Total-Variance Ratio:")
print(f"Var(X) / Var(Y) = {var_X}/{var_Y} = {gain:.6f}")

y_example = 5
print(f"\nIf Y = {y_example},")
print(f"Best estimate = E[X|Y=5] = {gain * y_example:.6f}")

# 4. RESIDUAL ANALYSIS


# X_hat = E[X|Y] = (1/5)Y
# Z = X - X_hat

print("\n" + "="*70)
print("4. VERIFICATION OF THE RESIDUAL")
print("="*70)
print("Residual:")
print("Z = X - E[X|Y] = X - (1/5)Y")
print()
print("E[Z] = E[X] - E[(1/5)Y] = 0 - 0 = 0")
print()
print("Cov(Z, Y) = Cov(X - (1/5)Y, Y)")
print("          = Cov(X,Y) - (1/5)Var(Y)")
print(f"          = {cov_XY} - {gain:.6f} × {var_Y}")
print("          = 0")
print()
print("Thus Z is uncorrelated with Y (Orthogonality Principle).")

# 5. DIFFERENTIAL ENTROPY

def differential_entropy(var):
    return 0.5 * math.log2(2 * math.pi * math.e * var)

h_X = differential_entropy(var_X)
var_X_given_Y = var_X - cov_XY**2 / var_Y
sigma_X_given_Y = np.sqrt(var_X_given_Y)

h_X_given_Y = differential_entropy(var_X_given_Y)

print("\n" + "="*70)
print("5. DIFFERENTIAL ENTROPY")
print("="*70)
print(f"Conditional variance Var(X|Y) = {var_X_given_Y:.6f}")
print()
print(f"h(X)      = {h_X:.6f} bits")
print(f"h(X|Y)    = {h_X_given_Y:.6f} bits")

# 6. MUTUAL INFORMATION


I_XY = h_X - h_X_given_Y
percent_removed = 100 * I_XY / h_X

print("\n" + "="*70)
print("6. MUTUAL INFORMATION")
print("="*70)
print(f"I(X;Y) = {I_XY:.6f} bits")
print(f"Percentage of uncertainty removed = {percent_removed:.2f}%")

# Case when noise variance = 100

var_Q_large = 100
var_Y_large = var_X + var_Q_large

gain_large = var_X / var_Y_large
var_cond_large = var_X - var_X**2 / var_Y_large

h_X_given_Y_large = differential_entropy(var_cond_large)
I_large = h_X - h_X_given_Y_large
percent_large = 100 * I_large / h_X

print("\nIf mechanical noise variance = 100:")
print(f"E[X|Y] = {gain_large:.6f} Y")
print(f"Var(X|Y) = {var_cond_large:.6f}")
print(f"I(X;Y) = {I_large:.6f} bits")
print(f"Uncertainty removed = {percent_large:.4f}%")

# 7. DIAGNOSTIC INTERPRETATION

print("\n" + "="*70)
print("7. DIAGNOSTIC INTERPRETATION")
print("="*70)
print("Manager's estimator: X_hat = Y")
print()
print("Why this is problematic:")
print("Y = X + Q includes mechanical noise Q.")
print("Therefore Y overreacts to noise and is not the MMSE estimator.")
print()
print("Optimal estimator:")
print("E[X|Y] = (1/5)Y")
print()
print("This shrinks the observation toward zero because:")
print("  - Only 1 unit of the total variance (5 units) is due to temperature.")
print("  - 4 units are due to noise.")
print("  - Therefore only 20% of Y should be attributed to X.")
print()
print("Example:")
print("If Y = 5")
print("  Manager estimate = 5")
print("  Optimal estimate = 1")
print()
print("Thus Y is an unbiased measurement of X plus noise,")
print("but it is not the best estimate of X given the observation.")


N = 200000

X_samples = np.random.normal(mu_X, sigma_X, N)
Q_samples = np.random.normal(mu_Q, sigma_Q, N)
Y_samples = X_samples + Q_samples

X_hat_samples = gain * Y_samples

Z_samples = X_samples - X_hat_samples

emp_mean_Z = np.mean(Z_samples)
emp_cov_ZY = np.cov(Z_samples, Y_samples)[0, 1]
emp_MSE = np.mean(Z_samples**2)

print("\n" + "="*70)
print("MONTE CARLO VERIFICATION")
print("="*70)
print(f"Empirical E[Z]      = {emp_mean_Z:.6f}")
print(f"Empirical Cov(Z,Y)  = {emp_cov_ZY:.6f}")
print(f"Empirical MSE       = {emp_MSE:.6f}")
print(f"Theoretical MSE     = {var_X_given_Y:.6f}")


print("\n" + "="*70)
print("FINAL ANSWERS SUMMARY")
print("="*70)
print("(X,Y) ~ N( [0,0], [[1,1],[1,5]] )")
print("Y ~ N(0,5)")
print("Cov(X,Y) = 1")
print("E[X|Y] = (1/5)Y")
print("Var(X|Y) = 4/5")
print("Residual Z = X - (1/5)Y")
print("E[Z] = 0")
print("Cov(Z,Y) = 0")
print(f"h(X) = {h_X:.6f} bits")
print(f"h(X|Y) = {h_X_given_Y:.6f} bits")
print(f"I(X;Y) = {I_XY:.6f} bits")
print(f"Uncertainty Removed = {percent_removed:.2f}%")
print("If Var(Q)=100, information becomes very small and the estimate shrinks heavily.")
```
